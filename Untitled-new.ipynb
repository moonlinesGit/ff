```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
file_path = '/content/drive/My Drive/dataset.csv'
df = pd.read_csv(file_path)

# Assume timestamp column is named 'time'
timestamps = df['time'].sort_values().to_numpy()

# Compute inter-arrival times
inter_arrival_times = np.diff(timestamps)  # Δt_i
T = len(inter_arrival_times)
```

Step 2: Initialize Model Parameters

```python
K = 3  # Number of hidden states

# Randomly initialize lambdas (rates per second)
lambdas = np.random.uniform(1, 20, size=K)

# Initialize transition matrix (must be Q matrix)
Q = np.random.rand(K, K)
np.fill_diagonal(Q, 0)
Q = Q / Q.sum(axis=1, keepdims=True)
np.fill_diagonal(Q, -Q.sum(axis=1))
```

EM Algorithm (Simplified)

```python
from scipy.stats import expon
from scipy.linalg import expm

def forward_backward(inter_times, Q, lambdas, pi):
    N = len(inter_times)
    K = len(lambdas)
    log_alpha = np.zeros((N, K))
    log_beta = np.zeros((N, K))

    # Initial forward log-probabilities
    log_alpha[0] = np.log(pi) + np.log(expon.pdf(inter_times[0], scale=1/lambdas))

    # Forward
    for t in range(1, N):
        for j in range(K):
            trans_probs = expm(Q * inter_times[t-1])[:, j]
            emis = expon.pdf(inter_times[t], scale=1/lambdas[j])
            log_alpha[t, j] = np.log(np.sum(np.exp(log_alpha[t-1]) * trans_probs)) + np.log(emis)

    # Backward
    log_beta[-1] = 0
    for t in reversed(range(N-1)):
        for i in range(K):
            trans_probs = expm(Q * inter_times[t])  # P_Δt
            emis = expon.pdf(inter_times[t+1], scale=1/lambdas)
            log_beta[t, i] = np.log(np.sum(trans_probs[i] * np.exp(log_beta[t+1]) * emis))

    # Posterior (gamma)
    log_gamma = log_alpha + log_beta
    log_gamma -= log_gamma.max(axis=1, keepdims=True)
    gamma = np.exp(log_gamma)
    gamma = gamma / gamma.sum(axis=1, keepdims=True)

    return gamma
```

Train Parameters (E + M steps)

```python
def train_mmpp(inter_times, K=3, n_iter=20):
    N = len(inter_times)

    # Init
    lambdas = np.random.uniform(1, 15, size=K)
    Q = np.random.rand(K, K)
    np.fill_diagonal(Q, 0)
    Q = Q / Q.sum(axis=1, keepdims=True)
    np.fill_diagonal(Q, -Q.sum(axis=1))
    pi = np.ones(K) / K

    for epoch in range(n_iter):
        gamma = forward_backward(inter_times, Q, lambdas, pi)

        # M-Step: Update λ
        for k in range(K):
            numer = np.sum(gamma[:, k])
            denom = np.sum(gamma[:, k] * inter_times)
            lambdas[k] = numer / denom

        # (Optional) update Q using transitions (not included here for brevity)
        # You can reestimate Q using expectation of transitions

        print(f"Epoch {epoch+1}: λ = {np.round(lambdas, 2)}")

    return lambdas, Q
```

Train Your Model

```python
lambdas_learned, Q_learned = train_mmpp(inter_arrival_times, K=3, n_iter=25)

print("\nFinal Learned λ (per state):", np.round(lambdas_learned, 2))
print("Final Learned Q matrix:\n", np.round(Q_learned, 3))
```
